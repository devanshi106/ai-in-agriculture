```python
import pandas as pd
yield_data = pd.read_csv("../data/crop_yield.csv.zip")
yield_data.shape
yield_data.describe()
yield_data.info()
yield_data.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
crop_encoder = LabelEncoder()
season_encoder = LabelEncoder()
state_encoder = LabelEncoder()
yield_data['Crop'] = crop_encoder.fit_transform(yield_data['Crop'])
yield_data['Season'] = season_encoder.fit_transform(yield_data['Season'])
yield_data['State'] = state_encoder.fit_transform(yield_data['State'])
yield_data.head()
```

```python
x = yield_data.drop('Yield', axis=1)
y = yield_data['Yield']

print(x.head())
print(y.head())
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

print(x_train.shape)
print(x_test.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=50, random_state=42)
model.fit(x_train, y_train)
print("Model training completed.")
```

```python
from sklearn.metrics import mean_absolute_error, r2_score

predictions = model.predict(x_test)
mae = mean_absolute_error(y_test, predictions)

r2 = r2_score(y_test, predictions)
print(f"Mean Absolute Error: {mae}")
print(f"R^2 Score: {r2}")
```

```python
import joblib
joblib.dump(model, "../models/yield_model.pkl")
joblib.dump(crop_encoder, "../models/crop_encoder.pkl")
joblib.dump(season_encoder, "../models/season_encoder.pkl")
joblib.dump(state_encoder, "../models/state_encoder.pkl")
print("Model and encoders saved successfully!")
```
